```python


import pandas as pd
import pyodbc

# Load your CSV
df = pd.read_csv('C:/Users/vbsr0/Desktop/softwares and notes-20250719T171616Z-1-001/python projects/moderation_dashboard/data/moderation_raw.csv')

# SQL Server connection
server = r'Vir\SQLEXPRESS'
database = 'vso_dashboard'

conn = pyodbc.connect(
    f'DRIVER={{ODBC Driver 17 for SQL Server}};'
    f'SERVER={server};'
    f'DATABASE={database};'
    'Trusted_Connection=yes;'
)

cursor = conn.cursor()

# SQL Server uses ? instead of %s
insert_query = """
    INSERT INTO moderation_logs 
    (date, reviewer_id, content_category, shift, region, 
     decision_outcome, correct_decision, processing_time_mins, sla_breached)
    VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)
"""

for _, row in df.iterrows():
    cursor.execute(insert_query, (
        row['date'],
        row['reviewer_id'],
        row['content_category'],
        row['shift'],
        row['region'],
        row['decision_outcome'],
        int(row['correct_decision']),      # BIT = 0/1
        row['processing_time_mins'],
        int(row['sla_breached'])           # BIT = 0/1
    ))

conn.commit()

print(f" {cursor.rowcount} rows inserted")

cursor.close()
conn.close()
```
